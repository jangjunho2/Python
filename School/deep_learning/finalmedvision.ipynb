```python
# Google Colab 연동
from google.colab import drive
drive.mount('/content/drive')

# 필요한 라이브러리 import
import os #운영 체제 관련 (경로)
from tensorflow.keras.models import Sequential #모델 생성
from tensorflow.keras.layers import Dense, Dropout, Flatten #모델 층 생성
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping #콜백 함수
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array #이미지 증강
from keras.applications.mobilenet import MobileNet #전이 학습 시킬 모델
from tensorflow.keras import optimizers #모델 최적화
import matplotlib.pyplot as plt #그래프 출력
import numpy as np #수치 계산
from sklearn.model_selection import StratifiedKFold #kfold 교차 검증
import math

# 데이터 가져오기
!git clone https://github.com/jangjunho2/data.git

#학습할 횟수,earlystopping(콜백 함수) 설정 
# 런타임 -> 런타임 유형 변경 -> 하드웨어 가속기 GPU 키고 학습 
EPO= 10 #학습할 횟수
STOP_PAT=30 #20 이상 추천(과적합 추적 하고싶으면 20~30 이상 추천)



# 모델 생성 함수 정의
def model_fn():
    # MobileNet을 특성 추출기로 사용하는 모델 생성
    base_model=MobileNet(weights='imagenet',include_top=False,input_shape=(img_width,img_height,3)) #전이학습으로 전달할 모델
    model=Sequential() #모델 생성
    model.add(base_model) #전이 학습
    model.add(Flatten()) #평탄화 (1차원)
    model.add(Dense(32,activation='relu')) #32개 레이어 완전 연결 활성함수 relu
    model.add(Dropout(0.25)) #과적합 방지를 위해 전체 뉴런중 25%를 비활성화
    model.add(Dense(4,activation='softmax')) #출력 층 4개

    # 모델 컴파일 설정
    model.compile(loss='sparse_categorical_crossentropy', #다중분류에서 쓰이는 손실함수,레이블이 정수형(sparse)일때 사용
                  optimizer=optimizers.Adam(2e-5),metrics=['accuracy']) #학습률을 작은 값으로 설정하므로써 이미 전이 학습된 모델의 가중치를 크게 바꾸지 않으면서 세밀한 조정이 가능,(대신 더 많은 학습이 필요함)
    return model

# 이미지를 로드하고 증강하는 제너레이터
'''
  증강해서 생긴 새로운 이미지들은 메모리에 저장하여 사용 사진 파일로 저장하여 사용할 수 있으나 효율적인 사용을 위해 저장하지 않음
  이미지 증강: 고양이_1사진이 있고 옵션을 주면, 회전시킨 고양이_1 사진, 좌우대칭한 고양이_1이 사진 총 3개로 데이터가 늘어 나는것이 아니라 
               회전 할수도 있고 안할수도 있음, 대칭옵션도 대칭이 될수도있고 안될수도 있음 그러한 고양이_1 사진이 하나 생성됨
               만약 아무런 변화가 고양이_1 사진에 일어 나지 않았다고 하더라도 고양이_2 사진에서는 변화가 일어나 회전한 고양이 사진이 생성된다면
               다양한 변환 결과를 얻을수 있음
'''
def image_generator(file_paths, labels, batch_size):  #file_paths: np.array, labels: np.array, batch_size: int
    # 이미지 데이터 증강을 위한 제너레이터 생성
    datagen = ImageDataGenerator(
        rescale=1./255, #정규화 #0~255 사이인 픽셀 값들을 0~1사이 값으로 나타냄
        rotation_range=20, #회전
        width_shift_range=0.2, #가로 이동
        height_shift_range=0.2, #세로 이동
        zoom_range=0.2, #확대 축소
        fill_mode='constant' # 회전 가로 세로 이동을 하면서 생긴 여백을 검정색으로 채움
    )

    # 배치 단위로 이미지를 로드하고 증강하여 반환하는 무한 루프
    while True:
        index = np.random.permutation(len(file_paths)) #file_paths의 길이가 10이라면 0~9까지의 숫자를 겹치지 않게 랜덤으로 생성 #ex) index = [1 4 2 3 0 5 9 7 6 8]
        for i in range(math.ceil(len(file_paths)/batch_size)):# 올림(file_path의 길이 // batch_size )만큼 반복
            
            if (i+1)*batch_size>len(file_paths): # 마지막 루프에서 딱 나누어 떨어지지 않을때 처리
              batch_files = file_paths[index[i*batch_size:]] #0~(32-1)번째 파일 이름들을 담고 있는 np.array 그 이후  32~(64-1), 64~(96-1) .. 반복
              batch_labels = labels[index[i*batch_size:]] #0~(32-1)번째 파일 정답들을 담고 있는 np.array 그 이후  32~(64-1), 64~(96-1) .. 반복
            
            else:
              batch_files = file_paths[index[i*batch_size:(i+1)*batch_size]] #0~(32-1)번째 파일 이름들을 담고 있는 np.array 그 이후  32~(64-1), 64~(96-1) .. 반복
              batch_labels = labels[index[i*batch_size:(i+1)*batch_size]] #0~(32-1)번째 파일 정답들을 담고 있는 np.array 그 이후  32~(64-1), 64~(96-1) .. 반복

            batch_images = []
            for file_path in batch_files: #batch_file : np.array
                img = load_img(file_path, target_size=(img_width, img_height))# 이미지 로드
                img = img_to_array(img) #로드한 이미지 -> 배열
                img = datagen.random_transform(img) #이미지 증강 #실제적인 이미지 증강이 일어나는 부분
                batch_images.append(img) #생성된 이미지 리스트에 추가
            batch_images = np.array(batch_images) #list -> np.array
            yield batch_images, batch_labels #yield는 쉽게 말해서 return이랑 비슷, 정확히는 iterator의 next()와 비슷한 개념으로 하나씩 반환 해줌 iterator 끝에 도달한후에 next() 함수를 호출 하면 에러가 나지만 generator는 에러를 알아서 처리해줌
            '''
            자세한 설명 GPT:
                위 코드는 제너레이터 함수 내에서 사용되는 문장입니다. yield 문은 제너레이터 함수에서 값을 반환하는 역할을 수행합니다.

                yield 문은 return 문과 비슷한 역할을 하지만, 중요한 차이점이 있습니다. return 문은 함수를 종료하고 값을 반환한 후 함수의 실행이 완전히 종료됩니다. 하지만 yield 문은 함수의 실행을 일시 중지하고 값을 반환한 후, 다음 호출 때 이전 상태를 기억하고 실행을 이어나갑니다.

                제너레이터 함수에서 yield 문을 사용하여 값을 반환하면, 호출자는 이 값을 하나씩 받아올 수 있습니다. 제너레이터는 yield 문을 만날 때마다 값을 반환하며, 호출자는 이를 순차적으로 받아올 수 있습니다. 호출자는 제너레이터 객체를 만들고 next() 함수를 호출하여 제너레이터의 다음 값을 가져옵니다.

                중요한 점은 제너레이터는 이터레이터(Iterator)의 한 종류이며, 값을 생성하는 동안 상태를 유지합니다. 제너레이터의 모든 값을 반환하고 더 이상 반환할 값이 없을 때, 호출자는 StopIteration 예외를 발생시킵니다. 이를 통해 제너레이터 함수는 동적으로 값을 생성하고 반환하면서 호출자와 협력적으로 작동할 수 있습니다.

                따라서, yield 문은 제너레이터 함수를 통해 값을 한 번에 하나씩 반환하고, 호출자는 next() 함수를 사용하여 값을 순차적으로 받아올 수 있습니다. 이를 통해 데이터를 배치 단위로 동적으로 생성하고 반환할 수 있습니다. 또한, 제너레이터는 StopIteration 예외를 자동으로 처리하여 더 이상 반환할 값이 없을 때 예외가 발생하지 않도록 합니다.
            '''
# 이미지 크기 설정
img_width=150
img_height=150

# 데이터셋 경로 설정
folder_names = ['asprin2', 'cepacol2', 'mucinex2', 'zicam2']
data_dir = 'data/train'


# 파일 경로와 레이블 생성
file_paths = []
labels = []
for i, folder_name in enumerate(folder_names): #['asprin2', 'cepacol2', 'mucinex2', 'zicam2'] # folder_names: list
    folder_path = os.path.join(data_dir, folder_name)  #ex)  data/train/asprin2
    file_list = os.listdir(folder_path) # file_list= [asprin2 사진들의 이름이 저장된 리스트 ex) asprin2.0.jpg, asprin2.1.jpg],   foleder_path: str
    file_paths.extend([os.path.join(folder_path, file) for file in file_list]) #리스트 컴프리헨션,extend로 한번에 넣음으로써 성능올림 밑에 문장이랑 같음
    '''
    for file in file_list:
      file_paths.append(os.path.join(folder_path, file))

    이런 모양
    filepaths=["data/train/asprin2.0.jpg" , "data/train/asprin2.1.jpg" , ..... , "data/train/zicam2.199.jpg" , "data/train/zicam.200.jpg" ]
    '''
    labels.extend([i] * len(file_list)) #정답 레이블 생성 i가 0인 것은 asprin2가 정답이라는 것을 나타내고, len(file_list)는 정답이 0인것의 갯수 만큼 만든것임 


file_paths = np.array(file_paths) #위에서 만든 file_path를 넘파이 배열로 만듬 똑같음(,(쉼표)만 빠짐 모양)  ex) ['data/train/asprin2.0.jpg' 'data/train/asprin2.1.jpg' ...]
labels = np.array(labels) #위에서 만든 정답 레이블 넘파이 배열로 만듬 ex) [0 0 0 ....3]

# 정확도 저장을 위한 리스트
accuracy_list_train = []
accuracy_list_val = []

# StratifiedKFold를 사용하여 데이터를 다섯 개의 폴드로 나누기
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=100)# 다섯개로 나눔 #섞기 O #랜덤시드값 100

# 각각의 폴드에 대해 학습과 검증을 수행
for fold, (train_index, test_index) in enumerate(skf.split(file_paths, labels)): # file_paths:np.array , labels:np.array
    '''
      겹치지 않는 랜덤 숫자
      train_index: np.array = [1 2 3 5 10 ...] 랜덤 숫자 
      train_index: np.array = [4 6 7 8 9 ...] 랜덤 숫자 
    '''
    train_files = file_paths[train_index]
    train_labels = labels[train_index]
    test_files = file_paths[test_index]
    test_labels = labels[test_index]
    
    print(f"Fold {fold+1}:")

    # 모델 생성
    model = model_fn()

    # 배치 크기 설정
    batch_size = 32 
    '''
      한번 학습할때 32개의 데이터 만큼 처리 , epochs가 책 회독수 라면, batch_size는 책 페이지 수로 한번에 학습할 페이지 
      한번볼때에 32p 만큼 학습해서 책 한권을 볼거고 책 한권 다 보면 epoch 1증가  
    '''

    # 훈련 및 테스트 데이터를 위한 제너레이터 생성
    train_generator = image_generator(train_files, train_labels, batch_size)
    test_generator = image_generator(test_files, test_labels, batch_size)
    
    # 체크포인트 콜백 설정
    model_checkpoint = ModelCheckpoint(
        filepath='./drive/MyDrive/saved_model/model_{epoch:02d}_{val_loss:.2f}.h5', #파일경로
        monitor='val_loss', #검증 손실을 기준으로 성능을 평가
        save_best_only=True, #위에서 평가한 값이 제일 좋을때만 저장함
        verbose=1 #진행 상황 출력(저장 되었다는 내용 보여짐)
    )
    earlystopping=EarlyStopping(monitor='val_loss',patience=STOP_PAT) # 검증 손실값이,10회간 개선이 없으면 조기종료

    # 모델 훈련
    history = model.fit(
        train_generator, #학습할 데이터 # train_generator : generator
        steps_per_epoch = math.ceil(len(train_files) / batch_size),  #한 에폭에서 필요한 반복 횟수
        epochs=EPO, #학습할 횟수
        validation_data=test_generator, #검증에 사용할 데이터
        validation_steps = math.ceil(len(test_files) / batch_size), #한 에폭에서 필요한 반복 횟수
        callbacks=[model_checkpoint,earlystopping]  # 콜백함수 추가
    )

     # 최적의 모델을 로드하여 정확도 값 저장
    max_train_accuracy = max(history.history['accuracy'])#학습셋 최고 정확도 일떄 값 저장
    max_val_accuracy = max(history.history['val_accuracy'])#검증셋 최고 정확도 일때 값 저장
    accuracy_list_train.append(max_train_accuracy) #위에서 저장한 값 리스트에 추가
    accuracy_list_val.append(max_val_accuracy) #위에서 저장한 값 리스트에 추가

    

    # 서브플롯 생성
    fig, axs = plt.subplots(1, 2, figsize=(15, 5))  # 가로로 두 개의 서브플롯 생성

    # 첫 번째 서브플롯 (정확도)
    axs[0].plot(history.history['accuracy'], marker='o') #학습셋 정확도
    axs[0].plot(history.history['val_accuracy'], marker='o') #검증셋 정학도
    axs[0].set_title(f'Fold {fold+1} Model Accuracy')  # 폴드 번호를 제목에 추가
    axs[0].set_ylabel('Accuracy') #y축 정확도
    axs[0].set_xlabel('Epoch') #x축 반복
    axs[0].legend(['Train', 'Test'], loc='upper left') #범례표시 왼쪽 위
    axs[0].grid() #그리드표시 #격자무늬

    # 두 번째 서브플롯 (손실)
    axs[1].plot(history.history['loss'], marker='o') #학습 손실
    axs[1].plot(history.history['val_loss'], marker='o') #검증 손실
    axs[1].set_title(f'Fold {fold+1} Model Loss')  # 폴드 번호를 제목에 추가
    axs[1].set_ylabel('Loss') #y축 손실 값
    axs[1].set_xlabel('Epoch')#x축 반복
    axs[1].legend(['Train', 'Test'], loc='upper left') #범례표시 왼쪽 위
    axs[1].grid() #그리드표시 #격자무늬

    # 그래프 표시
    plt.show()

    print('-------------------------------------------------------------------------------------')
    print(f"Fold{fold+1} max_train_accuracy: {max_train_accuracy}") #학습
    print(f"Fold{fold+1} max_val_accuracy: {max_val_accuracy}") #검증
    print('-------------------------------------------------------------------------------------')


# 평균값 계산
train_average=sum(accuracy_list_train)/len(accuracy_list_train) #학습셋 평균
validation_average=sum(accuracy_list_val)/len(accuracy_list_val) #검증셋 평균

# x축 값 (폴드 번호)
x = range(1, 5 + 1) #1~5

# 1행 2열의 서브플롯 생성
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# 학습 정확도 그래프
axes[0].bar(x, accuracy_list_train) # 바그래프 학습셋 출력
axes[0].axhline(train_average, color='r', linestyle='--', label='Train Average') #평균값 선
axes[0].set_xlabel('Fold') # X축 이름
axes[0].set_ylabel('Accuracy') #Y 축이름
axes[0].set_title('Train Accuracy per Fold') #제목
axes[0].set_xticks(x) #눈금 생성
axes[0].legend()#범례표시

# 검증 정확도 그래프
axes[1].bar(x, accuracy_list_val) # 바그래프 검증셋 출력
axes[1].axhline(validation_average, color='r', linestyle='--', label='Validation Average')
axes[1].set_xlabel('Fold')# X축 이름
axes[1].set_ylabel('Accuracy')#Y 축이름
axes[1].set_title('Validation Accuracy per Fold') #제목
axes[1].set_xticks(x)#눈금 생성
axes[1].legend() #범례표시

# 그래프 표시
plt.tight_layout() #그래프 간격조절
plt.show()

# 평균값 출력
print(f"Train Acurracy List: {accuracy_list_train}")
print(f"Model Train Average : {train_average}")
print(f"Validation Acurracy List : {accuracy_list_val}")
print(f"Model Validation Average : {validation_average}")


```

훈련 과정에서는 model.fit 함수에서 제너레이터를 사용하여 데이터를 제공합니다. model.fit 함수는 제너레이터로부터 데이터를 가져와서 학습을 수행합니다.

여기서의 동작 순서는 다음과 같습니다:

1. model.fit 함수가 호출되면 제너레이터가 생성됩니다.
2. 첫 번째 호출에서는 제너레이터가 시작되고, yield 문을 만나서 첫 번째 배치 데이터가 반환됩니다.
3. 반환된 배치 데이터를 사용하여 모델이 학습을 수행합니다.
4. 학습이 완료되면 다음 호출에서 yield 문을 만나서 다음 배치 데이터가 반환됩니다.
5. 반환된 배치 데이터를 사용하여 모델이 추가로 학습을 수행합니다.
6. 이 과정을 필요한 만큼 반복합니다.
7. 제너레이터가 더 이상 반환할 데이터가 없을 때, 즉 데이터셋의 끝에 도달했을 때 학습이 종료됩니다.
즉, 제너레이터는 필요한 만큼의 데이터를 동적으로 생성하고 반환하며, 모델은 반환된 데이터를 사용하여 학습을 수행합니다. 이를 통해 큰 데이터셋을 효율적으로 학습할 수 있습니다.

train_generator의 타입은 제너레이터(generator)입니다.

train_generator는 image_generator 함수에서 정의된 제너레이터(generator) 객체입니다. 제너레이터는 값을 생성하고 반환하는 함수로, yield 키워드를 사용하여 값을 한 번에 하나씩 반환합니다. 제너레이터는 이터레이터(iterator)의 한 종류로, 값을 생성하는 동안 상태를 유지하고, 호출자가 값을 순차적으로 받아올 수 있도록 합니다.

model.fit() 함수는 학습을 진행하는 동안 데이터를 제공하기 위해 제너레이터 객체를 사용합니다. 따라서 train_generator는 학습 데이터를 동적으로 생성하여 제공하는 역할을 수행합니다. model.fit() 함수는 train_generator에서 제공되는 배치 단위의 데이터를 사용하여 모델을 학습합니다.

따라서 train_generator는 제너레이터(generator) 객체이며, 학습 데이터를 생성하여 제공하는 역할을 합니다.

무한 루프에 빠지지 않는 이유는 model.fit() 함수의 steps_per_epoch와 validation_steps 매개변수를 설정하여 제너레이터(generator)가 정확한 에포크(epoch) 횟수에 도달하면 종료되기 때문입니다.

model.fit() 함수에서 steps_per_epoch 매개변수는 한 번의 에포크(epoch)당 제너레이터(generator)로부터 가져올 배치(batch)의 개수를 지정합니다. 마찬가지로 validation_steps 매개변수는 검증 과정에서 제너레이터로부터 가져올 배치(batch)의 개수를 지정합니다.

따라서 model.fit() 함수는 steps_per_epoch와 validation_steps에 설정된 배치 개수만큼 제너레이터로부터 데이터를 가져오고, 해당 배치 개수에 도달하면 종료됩니다. 이를 통해 무한 루프에 빠지지 않고 원하는 에포크 횟수만큼 학습과 검증이 수행됩니다.
